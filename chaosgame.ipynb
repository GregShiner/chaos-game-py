```python
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
from typing import Callable, Tuple, Final, List
import random
import progressbar
import imageio
```

```python
def find_midpoint(p1: Tuple, p2: Tuple) -> Tuple:
    return (p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2

equilateral_triangle = [
    (0, 0),
    (-0.5, -sqrt(3) / 2),
    (0.5, -sqrt(3) / 2),
]

isosceles_triangle = [
    (0.3, 0.5),
    (0, 0),
    (1, 0)
]

square = [
    (0, 0),
    (1, 0),
    (1, 1),
    (0, 1),
]

shape = equilateral_triangle
```

```python
# create plot
fig, ax = plt.subplots()

# plot the triangle with a solid line connecting the points
ax.plot(
    [p[0] for p in shape] + [shape[0][0]],
    [p[1] for p in shape] + [shape[0][1]],
    "k-",
)

ax.set_aspect("equal")

# save the plot to a file
plt.savefig("slideshow/triangle.png")
```

```python
# create plot
fig, ax = plt.subplots()

# plot the triangle with a solid line connecting the points
ax.plot(
    [p[0] for p in shape] + [shape[0][0]],
    [p[1] for p in shape] + [shape[0][1]],
    "k-",
)

ax.set_aspect("equal")
ax.set_axis_off()

ITERATIONS = 10000
FRAMES_TO_SKIP = 100

current_point = (0.15, -0.5)

bar = progressbar.ProgressBar(widgets=[
    ' [', progressbar.Timer(), '] ',
    progressbar.Bar(),
    ' (', progressbar.ETA(), ') ',
])

bar.start(ITERATIONS)
for i in range(ITERATIONS):
    # pick a random point from the triangle
    random_point = random.choice(shape)
    # find the midpoint between the current point and the random point
    current_point = find_midpoint(current_point, random_point)
    # plot the current point
    ax.plot(current_point[0], current_point[1], "k.", markersize=2)
    if i % FRAMES_TO_SKIP == 0:
        plt.savefig(f"slideshow/{ITERATIONS}dots/{i}.png")
    bar.update(i)

images = []
for i in range(0, ITERATIONS, FRAMES_TO_SKIP):
    image = imageio.v2.imread(f"slideshow/{ITERATIONS}dots/{i}.png")
    images.append(image)

imageio.mimsave(f"slideshow/{ITERATIONS}dots.gif", images, fps=30)
plt.savefig(f"slideshow/{ITERATIONS}dots.png")
```
